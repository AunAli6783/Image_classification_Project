## to rewrite the actual kaggle path of train and trst data in the .yaml file

```python
import yaml

# Load the YAML file into a dictionary
with open('/kaggle/working/AI_Project/car_dataset.yaml', 'r') as file:
    data = yaml.safe_load(file)

# Modify the data (for example, updating paths)
data['train'] = '/kaggle/working/AI_Project/dataset/images/train'
data['val'] = '/kaggle/working/AI_Project/dataset/images/val'

# Write the updated content back to the file
with open('/kaggle/working/AI_Project/car_dataset.yaml', 'w') as file:
    yaml.dump(data, file)
```
